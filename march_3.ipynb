```python
"""Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot."""
#Ans: 
# 1. Bar Plot:

# Use: Bar plots are used to display and compare categorical data. They're effective for showing the distribution of
# a categorical variable and comparing values across different categories.
# Common Use Cases: Visualizing sales by product categories, comparing student performance by subject, analyzing 
# survey responses by different groups.

# 2.Histogram:

# Use: Histograms are used to display the distribution of a continuous numerical variable. They help visualize the
# underlying frequency or density distribution of the data.
# Common Use Cases: Understanding the distribution of ages in a population, visualizing the distribution of exam
#scores, analyzing the distribution of income levels.

# 3.Scatter Plot:

# Use: Scatter plots are used to display the relationship between two numerical variables. They're helpful in 
# identifying patterns, trends, and correlations between variables.
# Common Use Cases: Examining the relationship between temperature and ice cream sales, plotting height vs.
# weight, studying the correlation between study hours and exam scores.

# 4.Heatmap:

# Use: Heatmaps are used to display a 2D dataset where the values are represented by colors. They're useful for 
# visualizing relationships and patterns in large datasets.
# Common Use Cases: Visualizing correlation matrices, showing the intensity of sales across different time periods
# and product categories, displaying confusion matrices in machine learning.

# 5.Violin Plot:

# Use: Violin plots combine elements of box plots and kernel density plots to display the distribution of a
# numerical variable across different categories. They show the distribution shape as well as summary statistics.
# Common Use Cases: Comparing the distribution of exam scores among different schools, visualizing the
# distribution of salaries across job roles, comparing the distribution of ages among different groups.
```

```python
""" Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions."""
#Ans:
import seaborn as sns

fmri = sns.load_dataset('fmri')
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)
```

```python
""" Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'."""
#ANs:
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
plt.subplot(1, 2, 1)
sns.boxplot(x = 'pclass', y = 'age', data = titanic)

plt.subplot(1, 2, 2)
sns.boxplot(x = 'pclass', y = 'fare', data = titanic)
```

```python
""" Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset."""
#Ans:
diamonds = sns.load_dataset('diamonds')
sns.displot(x ='price', data = diamonds, hue ='cut',  multiple="stack")
```

```python
""" Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset."""
#ANs:
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue = 'species')
```

```python
""" Que 6: Use the "flights" dataset from seaborn to plot a heatmap."""
#ANs:import seaborn as sns

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for heatmap
flights_pivot = flights_data.pivot_table(index='year', columns='month', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 6))  # Set the figure size
sns.heatmap(data=flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)

# Add a title and labels to the axes
plt.title("Passenger Count by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```
